```python
from deepface import DeepFace
import pandas as pd
import os
```

```python
directory = 'archive/test/surprise'
files = os.listdir(directory)
index = 0

while index < len(files):
    filename = files[index]
    objs = DeepFace.analyze(img_path = os.path.join(directory, filename), 
        actions = ['emotion'],
        enforce_detection=False)
    
    emotion_image_pairs[filename] = (objs[0]['dominant_emotion'], 'surprise')
    index += 1

```

```python
emotion_image_pairs = dict()
```

```python
def isoEmotion(emotion):
    preds = []

    for item in emotion_image_pairs.items():
        if item[1][1] == emotion:
            preds.append(item[1])

    return preds
```

```python
def emotionAccuracy(preds):
    correct = 0

    for pred in preds:
        if pred[0] == pred[1]:
            correct+=1

    return correct/len(preds)
```

```python
def calculate_precision_recall(true_positives, false_positives, false_negatives):
    precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) != 0 else 0
    recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) != 0 else 0
    return precision, recall
```

```python
angry_preds = isoEmotion('angry')
disgust_preds = isoEmotion('disgust')
fear_preds = isoEmotion('fear')
happy_preds = isoEmotion('happy')
neutral_preds = isoEmotion('neutral')
sad_preds = isoEmotion('sad')
surprise_preds = isoEmotion('surprise')
```

```python
angry_correct = emotionAccuracy(angry_preds)
disgust_correct = emotionAccuracy(disgust_preds)
fear_correct = emotionAccuracy(fear_preds)
happy_correct = emotionAccuracy(happy_preds)
neutral_correct = emotionAccuracy(neutral_preds)
sad_correct = emotionAccuracy(sad_preds)
surprise_correct = emotionAccuracy(surprise_preds)
```
